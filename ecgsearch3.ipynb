```python
# importer les bibliotheques 

import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
import statistics as st
from scipy.signal import find_peaks
from matplotlib.widgets import Button


```

```python
def frame_reading():
    dataFrame = pd.read_csv('ptbdb_normal.csv',delimiter=",")
    
    print(dataFrame)
    
   
    dataFrame_vide = pd.DataFrame()
    dataFrame_original = dataFrame
    dataFrame_original = dataFrame_original.T 
    dataFrame_vide = pd.concat([dataFrame_vide, dataFrame_original], axis=0)
    #print(dataFrame_vide)
    dataFrame_vide = dataFrame_vide.reset_index(drop=True)
    #print(dataFrame_vide)
    dataFrame_vide_nouv = dataFrame_vide.T
    #print(dataFrame_vide_nouv)
    columns_num = len(dataFrame.columns)
    #print(columns_num)
    
    mainlist = []
    for i in np.arange(0.000,columns_num/125, 0.008):
        mainlist.append(i)
    
    
    #print(mainlist)
    
    
    for index,row in dataFrame.iterrows():
        #print('the row is :\n',row)
        
        
        y= np.array(row).tolist()
        #print(y)
    #print(len(y))
    
    """
    columns = list(dataFrame_vide)
    for i in columns: 
  
        # printing the third element of the column 
        print (dataFrame_vide)

    """
    """
    for index,col in dataFrame.items():
        print(col)
        
    """  
    
    #fig=plt.figure(figsize=(17,6))
    plt.plot(mainlist,y,alpha=0.8,c="red",linewidth=2); plt.legend(['Signal ECG'])     
    plt.xlabel("Temps")
    plt.ylabel("Amplitude")
    #plt.grid()
    
    # importing all necessary libraries
    from itertools import count
    from matplotlib import animation
    from matplotlib.animation import FuncAnimation
    %matplotlib qt
    
    
    myvar = count(0)
    fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(17, 6))
    x1, y1 = [], []
    myvar = count(0)
 
    def animate(i):
        x1.append(next(myvar))
        y1.append((y[i]))
        
        axes.plot(x1, y1, color="red")
    
    # set ani variable to call the
    # function recursively
    anim = FuncAnimation(fig, animate, interval=10)
    
    """
    fig, ax =plt.subplots()
    p, = plt.plot(mainlist,y,alpha=0.8,c="red",linewidth=2); plt.legend(['Signal ECG'])  
    
    icone_ecg=plt.imread('C:/Users/KACED/Desktop/amine-youcef-pc/stage/ECG code/ecg.JPG')
    axButton1=plt.axes([0.05,0.01,0.1,0.1])
    btn1 = Button(ax=axButton1,
                  label='',
                  image=icone_ecg,
                  color='teal',
                  hovercolor='tomato'
    )
    
    
    def color_green(event):
        p.set_color('green')
        plt.draw()
        
    btn1.on_clicked(color_green)
    
    
    
    plt.show()
    """
    
    
    
      
    
     
    somme =np.sum(y)
    moyenne=np.mean(y)
    median=np.median(y)
    median_high=st.median_high(y)
    median_low=st.median_low(y)
    variance=st.variance(y)
    ecart_type=np.std(y)
 
    skewness = y.skew()
    zero_crossings = np.where(np.diff(np.signbit(y)))[0] #les valeurs qui passent par le zero
    zerocrossing_number = len(zero_crossings)
    
    
    # maxima : use builtin function to find (max) peaks
    #max_peak, _ = find_peaks(y,distance=55,height=6500)
    max_peak, _ = find_peaks(y,prominence=400,height=6500)
    
    
    
    
    
    
    # inverse  (in order to find minima)
    inv_data = 1/y
    
    # minima : use builtin function to find (min) peaks (use inversed data)
    min_peak, _= find_peaks(inv_data,distance=55)
    
    
    fig=plt.figure(figsize=(17,6))
    plt.plot(max_peak, y[max_peak], "vg"); plt.plot(y)
    plt.plot(min_peak, y[min_peak], "ob"); plt.plot(y)
    plt.legend(('max','min'))
    plt.grid()
    
    
    
    
    
   

    
    
    #print("\nles maxima sont\n",y[max_peak])
    list_max = y[max_peak].tolist()
    #print("\nles y de maxima sont\n",list_max)
    
    #print("\nles minima sont\n",y[min_peak])
    list_min = y[min_peak].tolist()
    #print("\nles y de minima sont\n",list_min)
    
    #print("--------------------------------")
    max_maxima = max(list_max)
    #print("le max des maxima est :",max_maxima)
    min_maxima = min(list_max)
    #print("le min des maxima est :",min_maxima)
    
    max_minima = max(list_min)
    #print("le max des minima est :",max_minima)
    min_minima = min(list_min)
    #print("le min des minima est :",min_minima)
    
   
     
    
    
   
    
   
   
     
    return pd.DataFrame({'somme':[somme],'moyenne':[moyenne],'variance':[variance],'ecart_type':[ecart_type],'zerocrossing_number':[zerocrossing_number]
    
                        ,'max_maxima':[max_maxima],'min_maxima':[min_maxima],'max_minima':[max_minima],'min_minima':[min_minima]
    
                        ,'median':[median],'median_high':[median_high],'median_low':[median_low],'skewness':[skewness]})
    
           
    
    

  
    
```
